```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
img = cv2.imread('../assets/paraboloid.png')
img_gray = cv2.cvtColor(img, cv2.IMREAD_GRAYSCALE)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
d = []
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img_gray[i, j] > 180:
            img_gray[i, j] = 255
        else:
            d.append([i, j])
d = np.array(d)

def rotate(data, angle):
    A = np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]])
    return np.array([A @ e for e in data])

def get_dark_pixels(img_path, threshold=180):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    dark_pixels = []
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i, j] < threshold:
                dark_pixels.append([i/img.shape[0], j/img.shape[1]])
    return rotate(dark_pixels, 3.0 * np.pi / 2.0)
d = get_dark_pixels('../assets/paraboloid.png')
```

```python
plt.imshow(img,)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(d[:, 0], d[:, 1], s=0.01)
ax.set_aspect(1)
plt.show()
```
